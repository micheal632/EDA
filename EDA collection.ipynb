## It is interesting and delightful to note that EDA (exploratory  data analysis is not just a basic to data interpretation but it's a backbone to data explanation.
               

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df['Total score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
df.head()
```

```python
df.info()
```

```python
df['math score'] = pd.to_numeric(df['math score'], downcast = 'float')
```

```python
df.info()
```

```python
df.describe()
```

```python
print('Average math score : {} '.format(np.mean(df['math score'])))
print('Average reading score is : {}  '.format(np.mean(df['reading score'])))
print('Average writing score is : {}'.format(np.mean (df['writing score'])))
```

```python
plt.figure(figsize= (14,6))
plt.subplot(1,4,1)
#plt.title('MATH SCORES')
sns.boxplot (y= 'math score', data = df, palette= 'spring')
plt.subplot(1,4,2)
#plt.title('READING SCORE')
sns.boxplot(y= 'reading score', data = df, palette = 'summer')
plt.subplot(1,4,3)
#plt.title('WRITING SCORES')
sns.boxplot(y= 'writing score', data = df, palette = 'viridis_r')
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
sns.barplot(x= 'test preparation course', y= 'math score', data = df , hue = 'gender', palette = 'summer')
#plt.title("Grade")
plt.subplot(1,3,2)
sns.barplot (x= 'test preparation course', y= 'reading score', data = df , hue = 'gender', palette = 'summer')
#plt.title('READING SCORE')
plt.subplot(1,3,3)
sns.barplot( x= 'test preparation course', y= 'writing score', data = df, hue = 'gender', palette = 'summer' )
#plt.title('WRITING SCORES')
```

```python
sns.pairplot(data =df, hue ='gender', plot_kws={'alpha':0.2}, palette = 'gnuplot2_r')
```

```python
df['lunch'].unique()
```

```python
plt.figure(figsize = (20,8))
plt.subplot (1,3,1)
#plt.title('MATH SCORE')
sns.barplot(x= 'test preparation course', y= 'math score', data = df , hue = 'lunch', palette = 'terrain_r')

plt.subplot(1,3,2)
sns.barplot (x= 'test preparation course', y= 'reading score', data = df , hue = 'lunch', palette = 'terrain_r')
#plt.title('READING SCORE')
plt.subplot(1,3,3)
sns.barplot( x= 'test preparation course', y= 'writing score', data = df, hue = 'lunch', palette = 'terrain_r' )
#plt.title('WRITING SCORES')
```

```python
df[(df['math score']> 90)& (df['reading score']>90) & (df['writing score']>90)]\
.sort_values(by= ['Total score'], ascending = False)
```

```python
plt.figure(figsize = (20,8))
plt.subplot (1,3,1)
#plt.title('MATH SCORE')
sns.barplot(x= 'race/ethnicity', y= 'math score', data = df , hue = 'gender', palette = 'Paired')

plt.subplot(1,3,2)
sns.barplot (x= 'race/ethnicity', y= 'reading score', data = df , hue = 'gender', palette = 'Paired')
#plt.title('READING SCORE')
plt.subplot(1,3,3)
sns.barplot( x= 'race/ethnicity', y= 'writing score', data = df, hue = 'gender', palette = 'Paired' )
#plt.title('WRITING SCORES')
```
